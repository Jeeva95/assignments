```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits
```

```python
digits.data.shape
```

# The data consist of 8*8  pixel image,which means that they are 64 dimensional. To understanding the realtionship between the points pCA projects to 2D

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
projected = pca.fit_transform(digits.data)
```

```python
digits.data.shape
```

```python
projected.shape
```

# Lets plot first two principal components

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5)
plt.xlabel('componenet1')
plt.ylabel('componenet2')
plt.colorbar()
```
