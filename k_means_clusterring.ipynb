# Importing dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

# Dataset

```python
x,y = make_blobs(centers = 4, n_samples = 1000)
```

```python
x.shape
```

```python
y.shape
```

```python
fig = plt.figure(figsize = (8,6))
plt.scatter(x[:,0],x[:,1], c= y)
plt.title('Dataset with 4 clustters')
plt.xlabel('First features')
plt.ylabel('second features')
plt.show
```

# Finding the optional number of clusters using the elbow method

```python
wcss_list = []
for i in range(1,11):
    kmeans = KMeans()
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1,11), wcss_list)
plt.title('The elbow method')
plt.xlabel('number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

# Training the k-means algorithm on the training dataset

```python
kmeans = KMeans()
y_pred = kmeans.fit_predict(x)
```

```python
print(accuracy_score(y,y_pred)*100)
```
